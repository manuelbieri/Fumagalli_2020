# Set up a valid model

Whenever the parameters are not valid, an error message will appear with the details about the failure.

```python
model = Fumagalli_Motta_Tarantino_2020.MicroFoundationModel(gamma=0.3)

# print a summary of the model
print(model)
```

# Plot outcomes of the model

## I. Timeline of Model

```python
visualizer = Fumagalli_Motta_Tarantino_2020.Visualize.Timeline(model)
visualizer.plot()
```

## II. Outcomes over a Range of Assets

```python
visualizer = Fumagalli_Motta_Tarantino_2020.Visualize.AssetRange(model)
visualizer.plot()
```

## III. Outcomes for different merger Policies

```python
visualizer = Fumagalli_Motta_Tarantino_2020.Visualize.MergerPoliciesAssetRange(model)
visualizer.plot()
```

# IV. Payoffs

```python
visualizer = Fumagalli_Motta_Tarantino_2020.Visualize.Payoffs(model)
visualizer.plot()
```
